```python
# cria um Dataframe

datas = {'nome': ['Argentina', 'Brasil', 'Franca', 'Italia', 'Reino Unido'],
         'continente': ['America', 'America', 'Europa', 'Europa', 'Europa'],
         'extensao': [2780, 8511, 644, 301, 244],
         'corVerde': [0,1,0,1,0]}

codes = ['AR', 'BR', 'FR', 'IT', 'UK']
countries = pd.DataFrame(datas, index=codes)

print(countries,'\n')

countries
```

```python
# Buscas mp DataFra,e

# Cria o DataFrame
datas = {'nome': ['Argentina', 'Brasil', 'Franca', 'Italia', 'Reino Unido'],
         'continente': ['America', 'America', 'Europa', 'Europa', 'Europa'],
         'extensao': [2780, 8511, 644, 301, 244],
         'corVerde': [0,1,0,1,0]}

codes = ['AR', 'BR', 'FR', 'IT', 'UK']
countries = pd.DataFrame(datas, index=codes)

# Testa se um dado rótulo de linha existe
has_BR = 'BR' in countries.index
has_US = 'US' in countries.index

print(f'Existe o rotulo BR? {has_BR}')
print(f'Existe o rotulo US? {has_US}')
print("---------------------------------")

# Testa se um dado rótulo de coluna existe

has_greenCollor = 'corVerde' in countries.columns
has_blueCollor = 'corAzul' in countries.columns

print(f'Existe a coluna corVerde? {has_greenCollor}')
print(f'Existe a coluna corAzul? {has_blueCollor}')
print("---------------------------------")

# Testa se o valor faz parte de uma coluna

has_Brasil = countries['nome'].isin(['Brasil'])
print("Existe Brasil na coluna nome? \n\n", has_Brasil)
print("---------------------------------")
```

```python
# Modificando o DataFrame

print('--- Dataframe Original ---')
print(countries)

# Inserindo o pais Japão
countries.loc['JP'] = {'nome':'Japao', 'continente':'Asia', 'extensao': 372, 'corVerde': 0}
print('\n--- Dataframe com o pais Japão ---')
print(countries)

# Altera a extensão do Brasil

countries.at['BR', 'extensao'] = 8512
print('\n--- Dataframe com a nova extensão do Brasil ---')
print(countries)

# Remove a Argentina e o Reino Unido
countries = countries.drop(['AR', 'UK'])
print('\n--- Dataframe sem a Argentina e o Reino Unido ---')
print(countries)

```
