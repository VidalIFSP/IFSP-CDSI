```python
import pandas as pd
```

```python
lojaA = pd.DataFrame({
    "loja": ["A", "A", "A"],
    "dia": ["sex", "sab", "dom"],
    "valor": [7500, 9500, 8200]
})

lojaB = pd.DataFrame({
    "loja": ["B", "B", "B"],
    "dia": ["sex", "sab", "dom"],
    "valor": [5100, 8250, 9900]
})

lojaC = pd.DataFrame({
    "loja": ["C", "C"],
    "dia": ["sab", "dom"],
    "valor": [7500, 11800]
})

lojaABC = pd.concat([lojaA, lojaB, lojaC], ignore_index=True)
print(lojaABC)
```

```python
d1 = pd.DataFrame({
    "carro": ["Hyundai", "Renault", "Fiat"],
})

d2 = pd.DataFrame ({
    "animal": ["Capivara", "Lobo", "Macaco"]
})

d3 = pd.concat([d1,d2], ignore_index=True)
print(d3)
```

```python
df_sql = pd.DataFrame({
    "email": ["elza.=_peixoto@astconsult.com.br", "carlos_pires@piscinasegura.com.br"]
})

df_python = pd.DataFrame({
    "email": ["elza.=_peixoto@astconsult.com.br", "roberto_moreira@gruporeis.net", "alice.nicole.barbisa@bb.com.br"]
})

# União (Relação de emails distintos)

alunos = pd.concat([df_sql, df_python], ignore_index=True)
print(alunos)

# Interseção (Quem fez ambos os cursos)
sql_e_python = df_sql.merge(df_python)
print(sql_e_python)

# Diferença (Quem fez só sql e fez só python)
sql_mas_nao_python = df_sql[df_sql.email.isin(df_python.email)==False]
print(sql_mas_nao_python)

# Diferença (Quem fez só python e fez só sql)
python_mas_nao_sql = df_python[df_python.email.isin(df_sql.email)==False]
print(sql_mas_nao_python)
```

```python
flags = pd.read_csv("/content/drive/MyDrive/ADS/DataScience/Source/flags.csv")
countries = pd.read_csv("/content/drive/MyDrive/ADS/DataScience/Source/countries.csv")

num_linhas_flags = flags.shape[0]
num_linhas_countries = countries.shape[0]

print(f"Número de linhas em flags: {num_linhas_flags}")
print(f"Número de linhas em countries: {num_linhas_countries}")

# Junção interna: paises comuns a ambas as bases
ambas = pd.merge(flags, countries, how="inner", left_on="name", right_on="country")
print(ambas)
num_linhas_ambas = ambas.shape[0]
print(f"Número de linhas em ambas: {num_linhas_ambas}")

#Left Join: paises apenas em flags
flags_only = pd.merge(flags, countries, how="left", left_on="name", right_on="country")
flags_only = flags_only[pd.isnull(flags_only['country'])==True]
num_flags_only = flags_only.shape[0]
print(f"Número de linhas em flags_only: {num_flags_only}")
print(flags_only)

#Right Join: paises apenas em coutries
countries_only = pd.merge(flags, countries, how="right", left_on="name", right_on="country")
countries_only = countries_only[pd.isnull(countries_only['country'])==True]
num_countries_only = countries_only.shape[0]
print(f"Número de linhas em countries_only: {num_countries_only}")
print(countries_only)
```
